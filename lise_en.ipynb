```python
from unidecode import unidecode
```

```python
def normalize_text(text):
    if isinstance(text, str):
        return unidecode(text).lower()
    else:
        return text
```

```python
for column in lise_frame.columns:
    lise_frame[column] = lise_frame[column].apply(normalize_text)
```
